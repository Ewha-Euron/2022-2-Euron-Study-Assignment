 그림1  그림2

그림 1은 이차원에서 distributional semantics를 통한 분포를 나타낸것이다. 그림 1에서와 같이 나라를 나타내는 단어들이 같은 공간에 모여있는 것을 확인할 수 있다. 그림 2에서는 비슷한 뜻을 가지는 come과 go 가 비슷한 위치에 모여있는 것을 확인할 수 있다. Being, been과 같이 비슷한 역할을 하는 단어끼리 모여있는 모습을 확인할 수 있다. 해당 예시들은 이차원에서 단어의 분포를 확인한 것이지만 300 차원에서 단어 벡터들의 분포를 통해 의미를 파악할 수 있다.

Word2vec

-매우 큰 텍스트 말뭉치가 있다.

-고정된 어휘 안에 있는 모든 단어는 벡터를 이용하여 표현된다.

-텍스트를 읽으면서 center 단어인 c와 context 단어인 o를 가진다.

-o와 c 가 서로에게 기여한 확률을 계산하기 위해 c와 o 벡터의 유사도를 이용한다.

-위의 확률을 최대화하기 위해 단어의 벡터를 계속 조정한다.

```python
import numpy as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

from sklearn.decomposition import PCA

import gensim.downloader as api
from gensim.models import KeyedVectors
```

gensim 패키지를 이용한다.

```python
model = api.load("glove-wiki-gigaword-100")
print(type(model))
```

```python
model['bread']
```

```python
model['croissant']
```

```python
model.most_similar('usa')
```

```python
model.most_similar('banana')
```

```python
model.most_similar(negative='banana')
```

```python
result= model.most_similar(positive=['woman', 'king'], negative=['man'])
print("{}: {:4f}".format(*result[0]))
```

```python
def analogy(x1, x2, y1):
    result = model.most_similar(positive=[y1, x2], negative=[x1])
    return result[0][0]
```

analogy 함수를 이용하여 입력된 단어들의 유사성을 출력할 수 있다.

```python
analogy('man', 'king','woman')
```

```python
analogy('king', 'man', 'queen')
```

```python
analogy('australia', 'beer', 'france')
```

```python
analogy('pencil', 'sketching', 'camera')
```

```python
analogy('obama', 'clinton', 'reagan')
```
